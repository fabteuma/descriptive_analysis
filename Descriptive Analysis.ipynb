## Descriptive Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/zip.train", header=None, sep=" ")
```

```python
data
```

```python
image=pd.DataFrame(data.iloc[2,1:257])
```

```python
image_reshaped=image.values.reshape((16,16))
```

```python
cleaned_data = data.iloc[:,:-1]
```

```python
cleaned_data.head()
```

```python
plt.imshow(image_reshaped, cmap='hot', interpolation = 'NONE')
plt.show()
```

```python
number_6=pd.DataFrame(cleaned_data[cleaned_data[0] == 6.0])
```

```python
mean_image=pd.DataFrame(number_6.iloc[:,1:].mean())
```

```python
mean_image_reshaped = mean_image.values.reshape((16,16))
```

```python
median_image=pd.DataFrame(number_6.iloc[:,1:].median())
```

```python
median_image_reshaped = median_image.values.reshape((16,16))
```

```python
quantile90_image=pd.DataFrame(number_6.iloc[:,1:].quantile(0.9))
```

```python
quantile90_image_reshaped = quantile90_image.values.reshape((16,16))
```

```python
quantile10_image=pd.DataFrame(number_6.iloc[:,1:].quantile(0.1))
```

```python
quantile10_image_reshaped = quantile10_image.values.reshape((16,16))
```

```python
plt.imshow(mean_image_reshaped, cmap='hot', interpolation = 'bilinear')
plt.show()
```

```python
plt.imshow(median_image_reshaped, cmap='hot', interpolation = 'bilinear')
plt.show()
```

```python
plt.imshow(quantile90_image_reshaped, cmap='hot', interpolation = 'bilinear')
plt.show()
```

```python
plt.imshow(quantile10_image_reshaped, cmap='hot', interpolation = 'bilinear')
plt.show()
```
